# Housing Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# 1. Read housing.header.txt as a dataframe.
df = pd.read_csv('housing.header.txt')
display(df)
# Report number of instances and features
num_instances = df.shape[0]
num_features = df.shape[1]
print(f"Number of instances: {num_instances}")
print(f"Number of features: {num_features}")
# Report all samples with respect to the Crim index on a plot (the x-axis shows the index of the sample, 
# and the y-axis shows the crim index of the sample).
plt.figure(figsize=(10, 6))
plt.scatter(df.index, df['Crim'], marker='o', color='blue', label='Crim')
plt.xlabel('Index of the sample')
plt.ylabel('Crim index')
plt.title('Crim Index of Samples')
plt.legend()
plt.show()
```

```python
#2.	Show both histogram of the Crim index and the density of the Crim index on a 1x2 frame (one row two columns).
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
# Histogram
sns.histplot(df['Crim'], bins=30, kde=False, ax=axes[0])
axes[0].set_title('Histogram of Crim Index')
axes[0].set_xlabel('Crim')
axes[0].set_ylabel('Frequency')

# Density plot
sns.kdeplot(df['Crim'], ax=axes[1])
axes[1].set_title('Density Plot of Crim Index')
axes[1].set_xlabel('Crim')
axes[1].set_ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
# 3. Show following four scatter plots in one frame (1x4), crim-medv, Rm-medv, Age-medv,Tax-medv. 
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# Scatter plot for Crim vs. Medv
sns.scatterplot(x='Crim', y='Medv', data=df, ax=axes[0])
axes[0].set_title('Crim vs. Medv')

# Scatter plot for Rm vs. Medv
sns.scatterplot(x='Rm', y='Medv', data=df, ax=axes[1])
axes[1].set_title('Rm vs. Medv')

# Scatter plot for Age vs. Medv
sns.scatterplot(x='Age', y='Medv', data=df, ax=axes[2])
axes[2].set_title('Age vs. Medv')

# Scatter plot for Tax vs. Medv
sns.scatterplot(x='Tax', y='Medv', data=df, ax=axes[3])
axes[3].set_title('Tax vs. Medv')

plt.tight_layout()
plt.show()
```

```python
# Explain how are they (Crim, Rm, Age, Tax) correlated to the medium house value (Medv)
```

- Crim vs. Medv: In towns where the crime rate is very low, near 0 the median hous value seems to be dependent on other factors. However, as Crim increases, the Medv tends to decrease, showing a negative correlation.
- Rm vs. Medv: There's a strong positive correlation. As the number of rooms increases, the median house value tends to increase as well.
- Age vs. Medv: There's no clear correlation that can be inferred from the graph.
- Tax vs. Medv: There's no clear correlation. From the graph we can see that there are distinct tax brackets within the data that have different median house values.

```python
# 6. Create a scatter plot between Rm and Crim and show all 506 properties on the plot. 
# Color property whose Medv value greater or equal to 24 as red, and the rest as blue.
plt.figure(figsize=(10, 6))
plt.scatter(df[df['Medv'] >= 24]['Rm'], df[df['Medv'] >= 24]['Crim'], color='red', label='Medv >= 24')
plt.scatter(df[df['Medv'] < 24]['Rm'], df[df['Medv'] < 24]['Crim'], color='blue', label='Medv < 24')
plt.xlabel('Rm')
plt.ylabel('Crim')
plt.title('Scatter Plot of Rm vs. Crim')
plt.legend()
plt.show()
```

```python
# 7. Report the pairwise correlation between every two variables (either as a matrix or as a level plot)
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Pairwise Correlation between Variables')
plt.show()
```

```python
# 8. Please explain which variable is mostly positively correlated to Medv (medium house value), 
# and which variable is mostly negatively correlated to Medv.
# variable mostly positively correlated with Medv
positive_correlation = correlation_matrix['Medv'].drop('Medv').idxmax()
highest_positive_corr = correlation_matrix.loc[positive_correlation, 'Medv']

# variable mostly negatively correlated with Medv
negative_correlation = correlation_matrix['Medv'].drop('Medv').idxmin()
lowest_negative_corr = correlation_matrix.loc[negative_correlation, 'Medv']

print(f"The variable mostly positively correlated with Medv is '{positive_correlation}' with correlation coefficient: {highest_positive_corr}")
print(f"The variable mostly negatively correlated with Medv is '{negative_correlation}' with correlation coefficient: {lowest_negative_corr}")
```

```python
# 9. Draw scatterplots to show relationship between each attribute and Medv, respectively.
attributes = df.columns.drop('Medv')  # Exclude Medv from attributes
num_attributes = len(attributes)
num_cols = 4  # Number of columns for subplots

num_rows = (num_attributes + num_cols - 1) // num_cols  # Calculate number of rows

plt.figure(figsize=(16, num_rows * 4))  # Adjust figsize based on number of rows

for i, attribute in enumerate(attributes, start=1):
    plt.subplot(num_rows, num_cols, i)
    plt.scatter(df[attribute], df['Medv'], alpha=0.5)
    plt.xlabel(attribute)
    plt.ylabel('Medv')

plt.tight_layout()
plt.show()
```

```python
# 10. Explain how to use scatterplots to find attributes which are positively correlated, negatively correlated, or independent of Medv, respectively.
```

1. Positive correlation: If the points tend to form an upward trend or pattern as one variable increases, it indicates a positive correlation between the two variables. (e.g. Medv vs. Rm)
2. Negative correlation: If the points tend to form a downward trend or pattern as one variable increases, it indicates a negative correlation between the two variables. (e.g. Medv vs. Lstat)
3. Independent:  If the point in the scatterplot appear randomly scattered without forming any discernible pattern, it suggests that the variables are independent of each other. In this case, changes in one variable do not systematically affect the other variable.
